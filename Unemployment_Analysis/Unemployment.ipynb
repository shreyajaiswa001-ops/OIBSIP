```python
region_unemployment = df.groupby('Region')[' Estimated Unemployment Rate (%)'].mean()

region_unemployment.plot(kind='bar')
plt.title('Average Unemployment Rate by Region')
plt.show()
```

```python
df[' Estimated Unemployment Rate (%)'].hist(bins=20)
plt.title('Distribution of Unemployment Rate')
plt.show()
```

```python
df[' Date'] = pd.to_datetime(df[' Date'])
df.groupby(' Date')[' Estimated Unemployment Rate (%)'].mean().plot()
plt.title('Unemployment Rate Over Time')
plt.show()
```

```python
import seaborn as sns

sns.heatmap(
    df[[' Estimated Unemployment Rate (%)',
        ' Estimated Employed',
        ' Estimated Labour Participation Rate (%)']].corr(),
    annot=True
)

plt.show()
```

1. Unemployment rates vary across regions.

2. Some regions have significantly higher unemployment.

3. The unemployment rate changes over time.

4. Employment and labour participation show relationships with unemployment
